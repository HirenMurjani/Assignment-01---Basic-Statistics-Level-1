# Question 22

```python
import scipy.stats as stats
confidence_intervals = [0.90, 0.94, 0.60]
for x in confidence_intervals:
    z_scores = stats.norm.ppf((1 + x) / 2)
    print(f"Z score for {x*100} % Confidence Interval: {z_scores:.3f}")
```

# Question 23

```python
import scipy.stats as stats

sample_size = 25
confidence_intervals = [0.95, 0.96, 0.99]

for x in confidence_intervals:
    df = sample_size - 1
    t_scores = stats.t.ppf((1 + x) / 2, df=df)
    print(f"T score for {x*100}% Confidence Interval: {t_scores:.3f}")
```
